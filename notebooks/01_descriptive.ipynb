# Descriptive Analysis

Overview of all study data: row counts, distributions, missing values.
Adjust the optional `study_id` / `experiment_id` arguments to filter to a specific study.

```python
# Section 1 — Setup & Data Loading
from utils.db import load_handover, load_eyetracking, load_participants, load_questionnaire_responses
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Load all domains — add study_id=N or experiment_id=N to filter
df_handover = load_handover()
df_et       = load_eyetracking()
df_p        = load_participants()
df_qr       = load_questionnaire_responses()

print('Row counts:')
print(f'  handover:              {len(df_handover)}')
print(f'  eye_tracking:          {len(df_et)}')
print(f'  participants:          {len(df_p)}')
print(f'  questionnaire_response:{len(df_qr)}')
```

```python
# Section 2 — Data Preparation
# Convert timestamp columns to datetime if not already parsed
for col in ['giver_grasped_object', 'receiver_touched_object',
            'receiver_grasped_object', 'giver_released_object']:
    if col in df_handover.columns:
        df_handover[col] = pd.to_datetime(df_handover[col], errors='coerce')

# Derive total handover duration in milliseconds
df_handover['duration_ms'] = (
    df_handover['giver_released_object'] - df_handover['giver_grasped_object']
).dt.total_seconds() * 1000
```

```python
# Section 3 — Missing Values Overview
print('Missing values per column (handover):')
print(df_handover.isnull().sum())
```

```python
# Section 4 — Distributions
df_handover['duration_ms'].dropna().plot.hist(bins=30, title='Handover Duration (ms)')
plt.xlabel('Duration (ms)')
plt.tight_layout()
plt.show()
```

```python
# Participant overview
print('Participant summary:')
print(df_p.describe())
```
